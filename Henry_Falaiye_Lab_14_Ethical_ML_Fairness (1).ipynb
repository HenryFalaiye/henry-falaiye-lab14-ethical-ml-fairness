```python
import pandas as pd

data = {
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "Income": [50000, 60000, 40000, 70000, 30000, 80000, 45000, 52000],
    "LoanAmount": [200, 250, 180, 300, 150, 350, 220, 210],
    "Credit_History": [1, 1, 0, 1, 0, 1, 1, 0],
    "Loan_Status": [1, 1, 0, 1, 0, 1, 1, 0]
}

df = pd.DataFrame(data)
df.head()
```

```python
X = df.drop("Loan_Status", axis=1)
y = df["Loan_Status"]
```

```python
sensitive_feature = df["Gender"]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X["Gender"] = le.fit_transform(X["Gender"])
sensitive_feature = le.fit_transform(sensitive_feature)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, sf_train, sf_test = train_test_split(
    X, y, sensitive_feature, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```
